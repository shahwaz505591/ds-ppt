    def generate_permutations(s):
    if len(s) <= 1:
        return [s]
    permutations = []
    for i in range(len(s)):
        char = s[i]
        remaining = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining):
            permutations.append(char + perm)
    return permutations

    if n < 0:
        raise ValueError("Square root is not defined for negative numbers")
    if n == 0:

Ans:def longest_common_prefix(strs):
    if not strs:
        return ""
    shortest = min(strs, key=len)
    for i, char in enumerate(shortest):
        for other in strs:
            if other[i] != char:
                return shortest[:i]
    return shortest

41. Write a Python program to find the smallest missing positive integer in a list.
Ans:def find_smallest_missing_positive(nums):
    n = len(nums)
    # Move all non-positive integers to the left side
    j = 0
    for i in range(n):
        if nums[i] <= 0:
            nums[i], nums[j] = nums[j], nums[i]
            j += 1
    
    # Consider only the positive integers
    positive_nums = nums[j:]
    
    # Mark the presence of a positive integer by negating the value at its index
    for num in positive_nums:
        index = abs(num) - 1
        if index < len(positive_nums):
            positive_nums[index] = -abs(positive_nums[index])
    
    # Find the first index with a positive value (missing positive integer)
    for i in range(len(positive_nums)):
        if positive_nums[i] > 0:
            return i + 1
    
    # If all positive integers from 1 to len(positive_nums) are present
    return len(positive_nums) + 1

# Example usage:
numbers = [3, 4, -1, 1]

42. Implement a function to find the longest palindrome substring in a given string.
Ans:def longest_palindrome_substring(string):
    n = len(string)
    longest_length = 0
    start = 0
    
    # Helper function to expand around center for odd-length palindrome
    def expand_around_center(s, left, right):
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        return right - left - 1
    
    for i in range(n):
        length1 = expand_around_center(string, i, i)  # Odd-length palindrome
        length2 = expand_around_center(string, i, i + 1)  # Even-length palindrome
        max_length = max(length1, length2)
        
        if max_length > longest_length:
            longest_length = max_length
            start = i - (max_length - 1) // 2
    
    return string[start:start + longest_length]

# Example usage:
text = "babad"
longest_palindrome = longest_palindrome_substring(text)
print(longest_palindrome)  # Output: "bab"
